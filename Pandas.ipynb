Podsumowując **Series** to praktycznie ta sama jednowymiarowa tablica numpy (**ndarray**) z dodatkiem edytowalnego indeksu.

Szybkie porównanie:

| Cechy                       | Pandas Series                              | NumPy ndarray                           |
|-----------------------------|--------------------------------------------|-----------------------------------------|
| **Typ danych**               | Może przechowywać dane różnych typów       | Zwykle przechowuje dane jednego typu    |
| **Indeksowanie**             | Ma etykiety (indeksy) dla każdego elementu | Indeksowanie tylko za pomocą pozycji    |
| **Operacje**                 | Operacje są wyrównywane na podstawie indeksów | Operacje są wykonywane element po elemencie według pozycji |
| **Obsługa brakujących danych**| Obsługuje brakujące dane (`NaN`)           | Obsługa tylko dla danych typu float |
| **Wydajność**                | Zbudowane na bazie NumPy, więc podobna szybkość | Bardzo zoptymalizowane do danych numerycznych |
| **Metadane**                 | Zawiera etykiety (indeksy)                 | Brak metadanych (czysty numeryczny zbiór danych) |


## DataFrame (Ramka Danych)
Podstawowy obiekt którym będziemy się bawić w Pandas. Jest tym samym czym ndarray jest dla numpy, czyli sercem całem biblioteki. Przechowywać w niej będziemy różne dane tabelaryczne, gdzie każdy rząd nazywany będzie **Observation** (obserwacja), lub **Sample** (próbka), natomiast kolumna nazywana jest **Feature** (cecha), rzadziej **Variable** (zmienna).

````python
df = pd.DataFrame({'Imie': ['Bartek', 'Kamil', 'Miłosz', 'Patrycja'],
                   'Wiek': [25, 26, 23, 26],
                   'Wzrost': [170., 181.5, 177., 160.]})
````

Pole **colunns** dataframe to dodatkowy atrybut/pole przechowujący nazwy kolumn. Jest to obiekt **index** więc możemy odwoływać się do poszczególnych kolumn po ich nazwie, podobnie jak z indeksami.

Jak wcześniej wspominałem, DataFrame to kolekcja obiektów **series** więc z każdego z nich można utworzyć nową kolumnę.

````python
bd_series = pd.Series({0: '17.01.1999',
                         1: '24.04.1998',
                         2: '12.06.2001',
                         3: '24.03.1998',
                       })
````

Pandas wykorzystuje numpy'owe typy **datetime64** and **timedelta64** o których oczywiście (jeszcze) nie wspomniałem.

````python
df['Data Urodzenia'] = pd.to_datetime(bd_series, dayfirst = True)
````

````python
df
````

````python
df['Waga'] = pd.Series({0: 76,
                         1: 97,
                         2: 76,
                         3: 62,
                       })
df
````

Nowe kolumny możemy również tworzyć na podstawie innych kolumn.

````python
df['BMI'] = df['Waga'] / (df['Wzrost'] / 100)**2 # Broadcasting operacji matematycznych działa tak samo jak w numpy
df
````

Można też kolumny usuwać za pomocą metody **drop(columns = [kolumny do usunięcia], inplace = False)** gdzie za pomocą argumentu columns wskazujemy które kolumny chcemy usunąć, a inplace czy DataFrame zostanie zapisana czy nie (False - nie, default)

## Wartości brakujące
Wartości brakujące w pandas zaznaczane są wartością **np.nan** z numpy ,lub też pythonowkim **None**. Jest wiele sposobów radzenia sobie z wartościami brakującymi jak imputacja, interpolacja, usuwanie kolumn zawierających bardzo dużą ilość brakujących danych. O tym będziemy mówić przy okazji pracy na prawdziwym datasecie.

````python
# szybkie tworzenie testowage dataframe i zastąpienie losowych wartości, wartościamy np.nan
df_missing_vals = pd.DataFrame(np.random.randn(1000, 10), columns = [f'Col_{i}' for i in range(1, 11)]) # np.random.randn zwraca 2d tablice losowych liczb próbkowanych z dystrybucji nornalnej (normal distribution)
random_x_indicies, random_y_indicies = np.random.randint(0, df_missing_vals.shape[0], 100), np.random.randint(0, df_missing_vals.shape[1], 100)
random_x_indicies, random_y_indicies
````

````python
df_missing_vals.values[(random_x_indicies, random_y_indicies)] = np.nan
df_missing_vals
````

Metoda **.isnull()** zwraca maskę w której zaznaczone są wartości brakujące. Następnie są sumowane przez metodę **sum()**. Następnie dzielimy przez ilość rzędów(obserwacji) i mnożymy przez 100, w skrócie liczby procentowy brak każdej kolumny.

````python
df_missing_vals.isnull().sum()
````

Liczymy % brakijących wartości w całym datasecie. Najpierw sumujemy wartości brakujące w rzędach, potem je sumujemy a następnie dzielimy przez rozmiar df i mnożymy razy 100. W wyniku dostajemy 1% wartości brakujących

````python
df_missing_vals.isnull().sum().sum() / (df_missing_vals.shape[0] * df_missing_vals.shape[1]) * 100
````

Wartości te możemy wypełnić jakąś wartością, np. średnią za pomocą metody **fillna()**.

````python
df_filled_mean = df_missing_vals.copy()
for col in df_filled_mean.columns:
  df_filled_mean[col] = df_filled_mean[col].fillna(df_filled_mean[col].mean())

((df_filled_mean.isnull().sum() / df_filled_mean.shape[0]) * 100)
````

Za pomocą metody **interpolate** możemy je wypełnić za pomocą interpolacji, czyli wypełniamy na podstawie pobliskich danych.

````python
df_interpolated = df_missing_vals.copy()
for col in df_interpolated:
  df_interpolated[col] = df_interpolated[col].interpolate()

((df_interpolated.isnull().sum() / df_interpolated.shape[0]) * 100)
````

Możemy również po prostu wyrzucić rzędy zawierające brakujące dane za pomocą **dropna()**. Subset wskazuje na podstawie jakich kolumn usuwamy rzędy.

````python
df_dropped_na = df_missing_vals.copy()
df_dropped_na = df_dropped_na.dropna(subset = ['Col_1'])
````

````python
len(df_missing_vals), len(df_dropped_na)
````

Często zamiast usuwać całe rekordy/obserwacje możemy usuwać same kolumny które mają zbyt dóżą ilość wartości brakujących.

````python
missing_percentage = df_missing_vals.isnull().sum() / len(df_missing_vals) * 100
columns_to_drop = missing_percentage[missing_percentage > 1]
columns_to_drop
````

````python
df_dropped_cols = df_missing_vals.copy()
df_dropped_cols.drop(columns = list(columns_to_drop.index), inplace = True)
df_missing_vals, df_dropped_cols
````
