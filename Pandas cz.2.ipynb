# Agregacja danych
Jest to niezbędna czynność procesu analizy danych i badań eksploracyjnych. Pozwala na sprawne obliczenie najważniejszych statystyk danych takich jak średnia, mediana, kwartyle itd. oraz ich grupowanie w celu dostrzeżenia ewentualnych trendów w danych.

# Podsumowanie funkcji agregujacych


| Funkcja      | Opis                                                                                       | Przykład                         |
|--------------|--------------------------------------------------------------------------------------------|----------------------------------|
| `sum()`      | Suma wartości w każdej kolumnie                                                            | `df['kolumna'].sum()`            |
| `mean()`     | Średnia wartość w każdej kolumnie                                                          | `df['kolumna'].mean()`           |
| `count()`    | Liczba niepustych wartości w każdej kolumnie                                               | `df['kolumna'].count()`          |
| `min()`      | Minimalna wartość w każdej kolumnie                                                        | `df['kolumna'].min()`            |
| `max()`      | Maksymalna wartość w każdej kolumnie                                                       | `df['kolumna'].max()`            |
| `std()`      | Odchylenie standardowe wartości w każdej kolumnie                                          | `df['kolumna'].std()`            |
| `var()`      | Wariancja wartości w każdej kolumnie                                                       | `df['kolumna'].var()`            |
| `median()`   | Mediana wartości w każdej kolumnie                                                         | `df['kolumna'].median()`         |
| `mad()`      | Średnie bezwzględne odchylenie od średniej (mean absolute deviation)                       | `df['kolumna'].mad()`            |
| `first()`    | Pierwsza wartość w każdej kolumnie                                                         | `df['kolumna'].first()`          |
| `last()`     | Ostatnia wartość w każdej kolumnie                                                         | `df['kolumna'].last()`           |
| `agg()`      | Możliwość zastosowania wielu funkcji agregujących na raz, w tym niestandardowych funkcji   | `df['kolumna'].agg(['sum', 'mean'])` |

# Grupowanie danych

Często podczas przetwarzania zbioru danych będziemy chcieli pogrupować nasze dane na podstawie różnych kolumn, np. jeśli są to kolumny kategoryczne (płeć, pełnoletni/niepełnoletni, paragon/faktura). Służy do tego metoda **groupby**. Dzieli ona dane na podstawie danych kolumn (jednej lub więcej) dzięki czemu możemy wykonywać różne operacje, jak poznane wcześniej agregacje tylko na danej 'grupie' danych.

Posłużę się tutaj zbiorem danych **tips** zawartego w module plotly.data.

```python
df_tips = plotly.data.tips(pretty_names = True)
df_tips.head()
```

Jak widać sama metoda nie zwraca DataFrame a obiekt DataFrameGroupBy. To właśnie na tym obiekcie wykonujemy pożądane operacje.

```python
df_tips.groupby('Payer Gender')
```

Możemy np. dowiedzieć się że Panowie średnio dają większe (nie o wiele ale nadal) napiwki. Realistycznie rzecz biorąc taką różnice można potraktować jako błąd statystyczny.

```python
df_tips.groupby('Payer Gender')['Tip'].mean()
```

Grupować możemy również po wielu kolumnach. Czyli palące panie dają więcej niż nie palące, a najwięcej 'napiwkują' niepalący panowie. Ale czy najczęściej?.

```python
df_tips.groupby(['Payer Gender', 'Smokers at Table'])['Tip'].mean()
```

Nie, jak widać jeśli liczymy na napiwek, unikamy dymu, albo ryzykujemy i idziemy do palących panów.

```python
df_tips.groupby(['Payer Gender', 'Smokers at Table'])[['Tip', 'Total Bill']].sum()
```

Możemy również wykorzystywać wiele funkcji agregacyjnych naraz lub nawet tworzyć własne. W celu wykorzystania ich na naszej grupę korzystamy z metody **agg()**, ponadto umożliwia ona użycie różnych funkcji dla różnych kolumn.

```python
df_tips.groupby('Party Size').agg({'Total Bill': ['sum', 'count'],
                                   'Tip': 'mean'})
```

Czyli najwięcej wydały pary, a średnio największy napiwek dają grupy 6-cio osobowe. Wracając do płci, zobaczmy ile % całego rachunku stanowią napiwki. Wykorzystamy do tego własną funkcję.

```python
grouped_tips = df_tips.groupby('Payer Gender').agg(
    Total_Bill_Sum=('Total Bill', 'sum'),
    Tip_sum=('Tip', 'sum')
)
grouped_tips
```

```python
grouped_tips['tip percentage'] = (grouped_tips['Tip_sum'] / grouped_tips['Total_Bill_Sum']) * 100
grouped_tips
```

Warto jeszcze wspomnieć że jeśli mamy do czynienia z cechami takimi jak płeć pandas obsługuje typ **Categorical**. Zmienne tego typu mogą przyjąć określoną ilość wartości np. Płeć - Mężczyzna, Kobieta, Diagnoza - Chory, Zdrowy itp. Ponadto typ ten może mieć określoną kolejność.

```python
df_tips['Payer Gender'] = df_tips['Payer Gender'].astype('category')
df_tips.dtypes
```

# Apply
Jest to kelejna bardzo przydatna metoda pozwalająca na 'zastosowanie' danej funkcji wzdłuż danej osi ramki danych.

```python
df = pd.DataFrame(np.arange(0, 100).reshape(20, 5), columns=[f'Col_{x + 1}' for x in range(5)])
df.head()
```

```python
def half(number):
  return number / 2

df.apply(half).head()
```

Jako że musimy przekazać funkcje, możemy oczywiście przekazać lambdę.

```python
df.apply(lambda x: x * 4).head()
```

Za pomocą argumentu **axis** określamy wzdłuż czego stosujemy funkcję. 0 - kolumny, 1 - rzędy.

```python
df.apply(lambda row: row.sum(), axis = 1)
```

## Cheat Sheet funkcji, działają tak samo jak w pythonie

| Metoda Pandas `str`     | Opis                                              |

| `str.title()`            | Zwraca string z wielką literą na początku każdego słowa. |

| `str.strip()`            | Usuwa białe znaki z początku i końca stringu.   |
| `str.lstrip()`           | Usuwa białe znaki z początku stringu.           |
| `str.rstrip()`           | Usuwa białe znaki z końca stringu.              |
| `str.replace()`          | Zastępuje podciąg innym podciągiem.              |
| `str.split()`            | Dzieli string na listę podciągów.               |
| `str.join()`             | Łączy elementy iterowalnego obiektu w jeden string. |
| `str.find()`             | Zwraca indeks pierwszego wystąpienia podciągu.   |
| `str.rfind()`            | Zwraca indeks ostatniego wystąpienia podciągu.   |
| `str.index()`            | Zwraca indeks pierwszego wystąpienia podciągu (wywołuje błąd, jeśli nie znaleziono). |
| `str.rindex()`           | Zwraca indeks ostatniego wystąpienia podciągu (wywołuje błąd, jeśli nie znaleziono). |
| `str.count()`            | Zlicza wystąpienia podciągu w stringu.          |
| `str.startswith()`       | Sprawdza, czy string zaczyna się od danego podciągu. |
| `str.endswith()`         | Sprawdza, czy string kończy się danym podciągiem. |
| `str.len()`              | Zwraca długość stringu.                         |
| `str.isalpha()`          | Sprawdza, czy wszystkie znaki w stringu są literami. |
| `str.isdigit()`          | Sprawdza, czy wszystkie znaki w stringu są cyframi. |
| `str.isspace()`          | Sprawdza, czy wszystkie znaki w stringu są białymi znakami. |
| `str.isalnum()`          | Sprawdza, czy wszystkie znaki w stringu są literami lub cyframi. |

Aby utworzyć tabelę przestawną wykorzystujemy funkcję **pivot_table**. Musimy podać obiekt DataFeame z którego chcemy ją utworzyć, wartości jakie do niej przekażemy, co będzie naszym nowym indeksem, co kolumnami i jaką funkcję agregującą wykorzystamy. **fill_value** określa czym są wypełnianie ewentualne wartości puste.

```python
pivot_table = pd.pivot_table(df_tips, values = 'Tip', index = 'Meal', columns = 'Party Size', aggfunc = 'mean', fill_value = 0)
pivot_table
```

```python
pivot_table = pd.pivot_table(df_tips, values = ['Tip', 'Total Bill'], index = 'Meal', columns = 'Party Size', aggfunc = ['mean', 'sum'], fill_value = 0)
pivot_table
```
