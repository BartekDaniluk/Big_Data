# Wykresy Liniowe (Line plots)
Wykresy na których wartości zmiennej przedstawiane są jako ciągła linia. Najłatwiej sobie to wyobrazić tak że każdy punkt danych jest ze sobą łączony, jak w książkach dla dzieci z ponumerowanymi kropkami

````python
import numpy as np

fig = plt.figure()
ax = plt.axes()

ax.plot(np.linspace(-10, 10, 1000), 1 / (1 + np.exp(-1*np.linspace(-10, 10, 1000))))
fig.show()
````

````python
import plotly.express as px
import plotly.graph_objects as go

x_val = np.linspace(-10, 10, 1000)
y_val = 1 / (1 + np.exp(-1 * x_val))
fig = px.line(x=x_val, y=y_val, title="Plotly: Wykres Liniowy")
fig.show()
````

dla najprostszych wykresów fig i ax można pominąć.

````python
plt.plot(np.linspace(0, 100, 1000), np.cos(np.linspace(0, 100, 1000)))
````

````python
x_val = np.linspace(0, 100, 1000)
fig = px.line(x=x_val, y=np.cos(x_val), title="Plotly: Cosinus")
fig.show()
````

Jeśli chcemy na jednym wykresie (Figure) umieścić wiele linii wystarczy że wywołamy funkcję **plot** wiele razy.

````python
plt.plot(np.linspace(0, 10, 100), np.cos(np.linspace(0, 10, 100)))
plt.plot(np.linspace(0, 10, 100), np.sin(np.linspace(0, 10, 100)))
plt.plot(np.linspace(0, 10, 100), np.sin(np.linspace(0, 10, 100)) / 2)
````

````python
x_val = np.linspace(0, 10, 100)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x_val, y=np.cos(x_val), mode='lines', name='cos(x)'))
fig.add_trace(go.Scatter(x=x_val, y=np.sin(x_val), mode='lines', name='sin(x)'))
fig.add_trace(go.Scatter(x=x_val, y=np.sin(x_val)/2, mode='lines', name='sin(x)/2'))
fig.update_layout(title="Plotly: Wiele linii")
fig.show()
````

````python
fig = plt.figure()
ax = plt.axes()

ax.plot(np.linspace(-10, 10, 100), (np.linspace(-10, 10, 100) ** 2) * 2)
ax.plot(np.linspace(-10, 10, 100), (np.linspace(-10, 10, 100) * 2 + 1) ** 2)
fig.show()
````

````python
x_val = np.linspace(-10, 10, 100)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x_val, y=(x_val**2)*2, mode='lines', name='y=2x^2'))
fig.add_trace(go.Scatter(x=x_val, y=(x_val*2+1)**2, mode='lines', name='y=(2x+1)^2'))
fig.update_layout(title="Plotly: Dwie krzywe")
fig.show()
````

## Zmiana stylu linii
Podając w metodzie argumenty **color** oraz **linestyle** możemy bawić się wyglądem linii.


````python
x = np.linspace(-10, 10, 1000)

plt.plot(x+1, color = 'red')
plt.plot(x+2, color = (1, 0, 1)) # akceptowanie jest RGB, wartości od 0 do 1
plt.plot(x+3, color = '#0bb052') # oraz hexcode
````

````python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=x+1, mode='lines', line=dict(color='red')))
fig.add_trace(go.Scatter(x=x, y=x+2, mode='lines', line=dict(color='rgb(255, 0, 255)')))
fig.add_trace(go.Scatter(x=x, y=x+3, mode='lines', line=dict(color='#0bb052')))
fig.update_layout(title="Plotly: Kolory linii")
fig.show()
````

````python
plt.plot(x+1, color = 'red', linestyle='dotted')
plt.plot(x+2, color = (1, 0, 1), linestyle='--')
plt.plot(x+3, color = '#0bb052', linestyle='-.')
````

````python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=x+1, mode='lines', line=dict(color='red', dash='dot')))
fig.add_trace(go.Scatter(x=x, y=x+2, mode='lines', line=dict(color='rgb(255, 0, 255)', dash='dash')))
fig.add_trace(go.Scatter(x=x, y=x+3, mode='lines', line=dict(color='#0bb052', dash='dashdot')))
fig.update_layout(title="Plotly: Style linii")
fig.show()
````

Możemy również połączyć te słowa kluczowe w 1 niezazwane.

````python
plt.plot(x, 'r-.')
plt.plot(x+1, 'g:')
plt.plot(x+2, 'b-')
````

````python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=x, mode='lines', line=dict(color='red', dash='dashdot')))
fig.add_trace(go.Scatter(x=x, y=x+1, mode='lines', line=dict(color='green', dash='dot')))
fig.add_trace(go.Scatter(x=x, y=x+2, mode='lines', line=dict(color='blue')))
fig.update_layout(title="Plotly: Skróty stylów")
fig.show()
````

## plot vs scatter
Główną różnicą między **plot** a **scatter** jest to że scatter jest o wiele potężniejszym narzędziem. W **scatter** każdy punkt może mieć własne cechy, nadawane na podstawie danych, co umożliwia wyraźniejsze zilustrowanie danych.

````python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c = colors, s=sizes, alpha = 0.3,
            cmap = 'viridis')
plt.colorbar();
````

````python
fig = go.Figure(data=[go.Scatter(
    x=x, y=y,
    mode='markers',
    marker=dict(
        size=sizes/20,
        color=colors,
        colorscale='Viridis',
        showscale=True,
        opacity=0.6
    )
)])
fig.update_layout(title="Plotly: Scatter ze zmiennym rozmiarem i kolorem")
fig.show()
````

# Wykresy Pudełkowe i Skrzypcowe (Box plot, Violin plot)
Bardzo przydane wykresy gdy chcemy zwizualizować dystrybucje cech danych, ze szczególną uwagą na medianę, kwartyle oraz potencjalne wartości odstające.
Zdecydowanie lepiej wyglądają w plotly ale o tym za chwilę 😉.

Skrzypcowy to praktycznie to samo co pudełkowy, tylko że zamiast 'pudełka' mamy **Kernel Density Estimate**, czyli sposób estymacji funkcji gęstości danej cechy (zmiennej losowej). Najprościej mówiąc pokazuje nam jak ta zmienna 'wygląda', tam gdzie jest szeroka jest wiele obserwacji, a im węższa tym jest ich mniej.

````python
plt.boxplot(tips[['total_bill', 'tip']], notch = True, labels = ['Total bill', 'Tip'])
````

````python
fig = go.Figure()
fig.add_trace(go.Box(y=tips['total_bill'], name='Total bill', boxpoints='all', notched=True))
fig.add_trace(go.Box(y=tips['tip'], name='Tip', boxpoints='all', notched=True))
fig.update_layout(title="Plotly: Boxplot")
fig.show()
````

````python
plt.violinplot(tips[['total_bill', 'tip']], showmeans = True, showmedians = True)
plt.xticks([1, 2], ['total_bill', 'tip'])
````

````python
fig = go.Figure()
fig.add_trace(go.Violin(y=tips['total_bill'], name='total_bill', box_visible=True, meanline_visible=True))
fig.add_trace(go.Violin(y=tips['tip'], name='tip', box_visible=True, meanline_visible=True))
fig.update_layout(title="Plotly: Violinplot")
fig.show()
````

#Wiele wykresów / pod-wykresy (subplots)
Jak na początku wspominałem obiekt **plt.Figure** może przechowywać wiele wykresów, a każdy z tych wykresów to obiekt **plt.Axes**, czyli żeby dodać wykres wystarczy dodać obiekt **axes** do **figure**.

````python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4])
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4])
ax1.plot(np.sin(np.linspace(0, 10, 100)))
ax2.plot(np.cos(np.linspace(0, 10, 100)))
````

````python
from plotly.subplots import make_subplots
fig = make_subplots(rows=2, cols=1)
fig.add_trace(go.Scatter(y=np.sin(np.linspace(0, 10, 100)), mode='lines', name='sin'), row=1, col=1)
fig.add_trace(go.Scatter(y=np.cos(np.linspace(0, 10, 100)), mode='lines', name='cos'), row=2, col=1)
fig.update_layout(title="Plotly: Subploty", height=600)
fig.show()
````

Jeśli nasze wykresy chcemy ułożyć w rzędy i kolumny możemy wykorzystać funkcję **plt.subplot**. Jako argumentu podajemy kolejno liczbę rzędów, kolumn, i na którym 'pod wykresie' jesteśmy.

````python
for plot_number in range(1, 5):
  plt.subplot(2, 2, plot_number)
  plt.text(0.5, 0.5, str((2, 2, f'wykres nr: {plot_number}')),
             fontsize = 18, ha = 'center')
````

````python
from plotly.subplots import make_subplots
fig = make_subplots(rows=2, cols=2, subplot_titles=[f'wykres nr: {i}' for i in range(1, 5)])
fig.update_layout(title="Plotly: Siatka subplotów", height=600)
fig.show()
````


  Albo jakimkolwiek innym języku

# HTML

W markdownie można również robić wstawki z html i css za pomocą  IPython.display, np.


````python
from IPython.display import display, HTML

banner_html = """
<div style="background-color: #4CAF50; color: white; padding: 20px; text-align: center; font-size: 24px;">
    🔥 Banner nie wygenerowany przez AI ponieważ znamy HTML🔥
</div>
"""

display(HTML(banner_html))
````

````python
from IPython.display import display, HTML

banner_html = """
<style>
    .banner {
        background: linear-gradient(45deg, #6a11cb, #2575fc); /* Gradient background */
        color: white;
        padding: 50px;
        text-align: center;
        font-size: 30px;
        border-radius: 15px;
        box-shadow: 0 10px 30px rgba(0, 0, 0, 0.2);
        animation: glow 1.5s ease-in-out infinite alternate;
    }

    .banner h1 {
        font-family: 'Arial', sans-serif;
        font-weight: bold;
        font-size: 40px;
    }

    .banner p {
        font-family: 'Arial', sans-serif;
        font-size: 20px;
    }

    .banner button {
        background-color: #ff4b5c;
        color: white;
        border: none;
        padding: 15px 30px;
        font-size: 18px;
        cursor: pointer;
        border-radius: 160px;
        transition: background-color 0.3s ease;
    }

    .banner button:hover {
        background-color: #ff1a2d;
    }

    @keyframes glow {
        0% {
            text-shadow: 0 0 5px rgba(255, 255, 255, 0.5), 0 0 10px rgba(255, 255, 255, 0.5);
        }
        100% {
            text-shadow: 0 0 20px rgba(255, 255, 255, 0.7), 0 0 30px rgba(255, 255, 255, 0.7);
        }
    }
</style>

<div class="banner">
    <h1>🔥 Przetwarzanie wielkich zbiorów danych🔥</h1>
    <p>Podczas którego 0% kodu jest generowane przez AI!</p>
    <button>Get Started(0 AI)</button>
</div>
"""

display(HTML(banner_html))
````
